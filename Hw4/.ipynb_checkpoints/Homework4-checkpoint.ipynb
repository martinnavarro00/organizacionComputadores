### Answers to the article questions

As seen in the graphs as the `bs` increased the time and mb/s increased but slighty from 256b to 1M but the line tended to increased when the `bs` went to 500M. When `bs` was 500M the performance seemed to decreased in both cases, SSD and External Hard Disk. This dramatic change is caused since at first we are working with low `bs` but then we are requiring the disks to use more resources when we change it to 500M.

For the `oflag` parameter we observed that the perfomance of the throughput decrease when it was off. This means that `oflag` allow the to write data faster or straight away, meaning there are no extra copies of duplicates.
